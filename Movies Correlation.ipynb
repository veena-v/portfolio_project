```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv('movies.csv')


```

```python
for cols in df.columns:
    missing = np.mean(df[cols].isnull())
    print(cols,missing,missing>0)
```

```python
df.head()
```

```python
df['votes'] = df['votes'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head(5)
```

```python
df['correct_year'] = df['released'].astype('str').str[-20:-16]
```

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget VS Gross earnings')
plt.xlabel('budget')
plt.ylabel('Gross earnings')
plt.show()
```

```python
df_new = df.copy()
```

```python
df_new.dtypes
```

```python
df_new['correct_year'] = pd.to_numeric(df_new['correct_year'], errors='coerce').astype('Int64')
```

```python
df_new.dtypes
```

```python
df_numerized = df_new
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dytpes=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df_new.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')

plt.show()
```

```python
correlation_mat =  df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
```

```python
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs > 0.5]
```
